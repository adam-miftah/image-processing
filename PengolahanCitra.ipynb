Konversi gambar ke original ke grayscale

```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('image.jpg')

# Konversi ke grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tampilkan hasil
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

Mendeteksi tepi menggunakan algoritma Canny

```python
# Deteksi tepi menggunakan Algoritma Canny
edges = cv2.Canny(gray_img, threshold1=100, threshold2=200)

# Tampilkan hasil
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')
plt.show()
```

Menerapkan gaussian blur

```python
# Terapkan Gaussian blur
blurred_img = cv2.GaussianBlur(gray_img, (5, 5), 200)

# Tampilkan hasil
plt.imshow(blurred_img, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```


Menyimpan gambar hasil pemrosesan

```python
cv2.imwrite('grayscale_image.jpg', gray_img)
cv2.imwrite('edges_image.jpg', edges)
cv2.imwrite('blurred_image.jpg', blurred_img)
```
